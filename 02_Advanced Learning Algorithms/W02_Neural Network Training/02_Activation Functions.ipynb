# Choosing Activation Functions
- It turns out that depending on what the target label or the ground truth label y is, there will be one fairly natural choice for the activation function for the output layer, and we'll then go and look at the choice of the activation function also for the hidden layers of our neural network.
- We can choose different activation functions for different neurons in our neural network, and when considering the activation function for different neurons in our neural network, and when considering the activation function for the output layer, it turns out there'll often be one fairly natural choice, dependign on waht is the target or the ground truth label y.
- Specifically, if we are working on a classification problem where y is either zero or one, so a binary classification problem, then the sigmoid activation function will almost always be the most natural choice.
- If we're solving a regression problem, in which y would be a numer that can be either +ve or -ve, in that case linear activation functinon is used.
- If y can only take a non-negative values, then the most natural choice will be ReLU activation function.
- So, in choosing the activation function to use for our output layer, usually depending on ehat label y we're trying to predict, there'll be one fairly natural choice
    
- For the hidden layer, it is recommended just using ReLU as a defualt activation function.
    Reason 1 : ReLU is a bit fater than sigmoid activation function.
    Reason 2 : The ReLU function goes flat only in one part of the graph, on the left is completely flat. Whereas the sigmoid activation function, it goes flat in 2 plcaes. It goes flat to the left of the graph and it goes flat to the right of the graph. If we're using gradient descent to train a neural network, then when we have a function that is flat in llot of places, gradient descent would be slow.
