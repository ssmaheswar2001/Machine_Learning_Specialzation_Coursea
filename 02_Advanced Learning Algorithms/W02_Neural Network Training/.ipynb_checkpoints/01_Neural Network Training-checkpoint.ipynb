# Training Details
- In general logistic regression : 
    1. Step 1 : Specifying how to compute the outputs given the input X and parameters.
    2. Step 2 : Specify loss and costs
    3. Step 3 : Minimize the cost function we trained logistic regression
    
- The same 3 steps is how we can train a neural network in Tensorflow.
    1. Step 1 : Specify how to compute the ouput given the input x and parameters W and B.
    2. Step 2 : To compile the model and to tell it what loss we want to use and here we used binary cross entropy loss function. Once we specify this loss taking an average over the entire training set also gives us the cost function for the neural network.
    3. Step 3 : To call function to try minimize the cost as a function of the parameters of the neural network.
    
- Step 1 : Specify how to compute the output given the input x and parameters w and b. This specifies the entire architecture of the neural network.
    
- Step 2: To specify what is the loss function and this will also define the cost function we use to train the neural network. Tensorflow knows that it costs us want to minimize is then the average, taking the average over all m training examples of the loss on all of the training examples. Optimizing this cost function will result in fitting the neural network to our binary classification problem. We can even use differen loss functions like squared error loss if regression problem is used.
    
-  Step 3 : Finally we will ask TensorFlow to minimize the cross-function. Using gradient descent to train the parameters of neural network, then we are repeatedly, for every layer l and for every unit j, update wlj according to wlj - (alphs)(partial derivative) and similar to the parameters b as well.
    - Inorder to use gradient descent, the key thing we need to compute is these partial derivate terms. What Tensorflow does and what is standard in neural network training, is to use an algorithm called backpropagation in order to compute these partial derivative term. It implements backpropagation all within this function called fit.
