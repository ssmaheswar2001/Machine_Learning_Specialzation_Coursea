#  Error metrics for skewed datasets
- If we're working on a machine learning application where the ratio of positive to negative examples is very skewed, very far from 50-50, then it turns out that the usual error metrics like accuracy don't work that well.
- 
- To evaluate a learning algorithm's performance with one rare clas it's useful to construct confusion matrix, which is a 2-by-2 matrix.
- Having divided the classification into these 4 cells, 3 common metrics we might compute are then the precision and recall. 
- The precision of the learnin algorithm computes of all the patients where we predicted y is equal to 1, what fraction actually has the rare disease. Inother words, precision is defined as the number of true positive divided by the number classified as positive, whart fraction did we acutally get right
- Recall : Of all patients that actually have the rare disease, what fraction did we correctly deteect as hiving it. Recall is defined as the number of true positive divided by the number of actual positive.
- 
- When we have a rare class, looking at precision and recall and making sure that both numbers are decently high that hopefully helps reassure us that our learning algorithm is acutually useful.

# Trading off precision and recall
- High precision would mean that if a diagnosis of patients have that rare disease, probably the patient does have it and it's accurate diagnosis.
- High recall means that if there's a patient with that rare disease, probably the algorithm will correctly identify that they do have that disease.
- By raising the threshold, we end up with higher precision, but lower recall.
- Lowering the threshold, lower precision and higher recall.
- 
- If we want to autoamatically trade-off precision and recall rather than have to do so ourself, there is another metric called the F1 score that is sometimes used to automatically combine precision recall to help pick the best value or the best trade-off b/w the two.
- In order to decide which algorithm to pick, it may be useful to find a way to combine precision and recall into a single score, so we can just look at which algorithm has the highest score.
- The most common way to combining precision recall is a compute called the F1 score, and the F1 score is a way to combining P and R but that gives more emphasis to whichever of these values is lower. The F1 score is a way of computing an average of sorts that pays more attention to whichever is lower.
- F1 score gives away to trade-off precision and recall.
- F1 score in math is called the harmonic mean of P and R.
