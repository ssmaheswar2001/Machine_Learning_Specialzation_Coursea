# Using Multiple Decision Trees
- One of the weakness of using a single decision tree is that the decision tree can be highly sensitive to small changes in the data. 
- One solution to make the algorithm less sensitive or more robust is to build not one decision tree, but to build a lot of decision tree and we call that a tree ensemble.
- The fact that changing just one training example causes the algorithm to come up with a different split at the root and therefore a totally different tree, that makes the algorithm just not that robust.
- Tree esemble which just means a collection of multiple trees.
- 
- If we has the ensemble trees, each one of these is maybe a plauible way to classify cat verus not cat. If we had a new test example that we wanted to classify, then what we would do is run all three of these trees on our new example and get them to vte on whether it's the final prediction
    
- The reason we use an ensemble of trees is by having lots of decision trees and having them vote, it makes our overall algorithm less sensitive to what any single tree may be doing because it gets many different votes and it makes our overall algorithm more robust.

# Random Forest Algorithm
- Now that we have a way to use sampling with replacement to create new training sets that are a bit similar to but also quite different from the original training set.
- Random forest algorithm which is one powerful tree ensamble algorithm that works much better than using a single decision tree.
    - We can generate an ensemble of trees. If we are given a training set of size M, then for B =1 to B so we do this B times.
    - We can use something with replacement to create a new training set of size M. So if we have 10 training examples and then we would train a decision tree on this data set. 
    - We then again use replacement to generate another training set of M or 10 training examples. We then train the decision tree on this new dataset and we end up with a somewhat different decision tree. Amd we may do this of B times.
    - We would then when we're trying to make a prediction, get these trees all votes on the correct final prediction.
    
- This specific instance creation of tree ensemble is sometimes also called a bagged decision tree.
- There's one modification to this that will actually make it work even much better and that changes the algorithm the bagged decision tree into the random forest algorithm. The key idea is that even with this sampling with replacement procedure sometimes we end up with always using the same split at the root node and very similar splits near the root note. So there's one modification to the algorithm to further try to randomize the feature choice at each node that can cause the set of trees and we learn to become more different from each other. so whan we vote them, we end up with an even more accurate prediction. The way this is typically done is at every note when choosing a feature to use to split if end features are available.
- We pick a random subset of K less than N features and allow the algorithm to choose only from the subset of K features. So, we would pick K features as the allowed features and then out of those k features the one with the highest information gain as the choise of feature to use the split.

# XGBoost
- Most commonly used way or implementatoom of decision tree ensembles or decision trees there's an algorithm called XGBoost.
- There's a modification to the back decision tree algorithm that can make it work much better
    - Given the training set of size m, we repeat B times, use smapling with replacement to create a new training set of size M and then train the decision tree on the new data set.
    - The first time throgh the loop, we may create a training set like that and train a decision tree like that.
    - Other than the first time, when sampling instead of picking from all m examples of equal probability with one over m probability, let's make it more likely that we'll pick misclassified examples that the previously trained trees do poorly on.
    
- We're goint to look at the decision trees, we've trained so far and look at what we're still not yet doing well on. And then when building the next decision tree, we're going to focus more attention on the examples that we're not yet doing well. So rather than looking at all the training examples, we focus more attention on the subset of examples is not yet doing well on and get the new decision tree, the next decision tree reporting ensemble to try to do well on them. And this is the idea behind boosting and it turns out to help learning algorithm learn to do better more quickly
- 
- Most widely used is XGBoost which stands for extreme gradient boosting.
- XGBoost has a good choice of the default splitting criteria and criteria for when to stop splitting. And one of the innovations in XGBoost is that is also has built in regularization to prevent overfitting.
    
- On note, rather than doing sampling with replacement XGBoost actually assigns differtn weights to different training examples. So it doesn't actually need to generate a lot of randomly chosen training sets and this makes it even a little bit more efficient than using a sampling with repacement procedure.

# When to use decision trees
- When should we pick tree ensembles or neural networks?
- Decision trees and tree ensembles will often 
    - work well on tabular data, also called structured data.
    - Not recommended for unstructured data (images, audio, text), neural network works better for unstructured data task.
    - If we are training just a single decision tree and that decision tree has only say a few dozen nodes we may be able to print out a decision tree to understand exactly how it's making decisions.
- One huge advantage of decision trees and tree ensembles is that they can be very fast to train.
- Neural networks
    - Works well on all types of data, including tabular (structed) and unstructured data.
    - Neural networks may be slower than a decision tree.
    - It works with transfer learning
- Even when we string together multiple neural networks we can train them all together using gradient descent. Whereas for decision trees we can only train one decision tree at a time.

# Lab : Tree Ensembles
- Use pandas to perform one-hot encoding of a dataset.
- Use scikit-learn to implement a Decision Tree, Randsom Forest and XGBoost models

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

RANDOM_STATE = 55
```

### Dataset : Heart Failure Prediction Dataset
- This data set contains 11 features that can be used to predict possible heart disease.
- Let's train a machine learning models to assist with diagnosing this disease.

```python
# Load the dataset usin pandas
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
df.shape
```

- We must perform some data engineering before working with the models. There are 5 categorial features. so we will use pandas to one-hot encode them.

### One-hot encoding using pandas
- First we will remove the binary variables, because one-hot encoding them would do nothing to them. To achive this we will just count how many different values there are in each categorical variable and consider only the variables with 3 or more values

```python
cat_variables = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
```

- As a reminder, one-host encoding aims to transform a categorical variables with n outputs into n binary variables.
- Pandas has a built-in method to one-hot encode variables, it is the fuction pd.get_dummies. There are several arguments to this fucntion, but here we will use only a few. They are:
    1. data : DataFrame to be used
    2. prefix : A list with prefixed, so we know which value we are dealing with
    3. columns : The list of columns that will be one-hot encoded. 'prefix' and 'columns' must have the same length.

```python
# This will replace the columns with one-hot encoded ones and keep the columns outside 'columns ' argument as it is.
df = pd.get_dummies(data = df,
                    prefix=cat_variables,
                   columns=cat_variables,
                   dtype=int)
```

```python
df.head()
```

- Choose the variables that will be the input features of the model.
    - The target is HearDisease
    - All other variables are features that can potentially be used to predict the target, HearDisease

```python
# Removing the target variable
features = [x for x in df.columns if x not in 'HeartDisease']
```

- We started with 11 features. Let's see how many feature variables we have after one-hot encoding

```python
len(features)
```

### Splitting the dataset

```python
X_train, X_val, y_train, y_val = train_test_split(df[features], df['HeartDisease'], train_size=0.8, random_state= RANDOM_STATE )
```

```python
print(f'Train Samples : {len(X_train)}')
print(f'Validation Samples : {len(X_val)}')
print(f'Target proportion : {sum(y_train)/len(y_val):.4f}')
```

## Building the Models
### Decision Trtee
- There are several hyperparameters in the Decision Tree object from scikit-learn.
- The hyperparameters we will use are:
    1. min_samples_split : The min number of samples required to split an internal node.
        Choosing a higher min_samples_split can reduce the num of splits and may help to reduce overfitting.
    2. max_depth : The max depth of the tree. 
        Choosing a lower max_depth can reduce the num of splits and may help to reduce overfitting

```python
min_samples_split_list = [2, 10, 30, 5, 100, 200, 300, 700] # If the number is an integer, then it is actual quantity of samples.
max_depth_list = [1, 2, 3, 4, 8, 16, 32, 64, None] # None means that there is not depth limit
```

```python
accuracy_list_train = []
accuracy_list_val = []
for min_samples_split in min_samples_split_list:
    # We can fit the model at the same time we define it, because the fit function returns the fitted estimator.
    model = DecisionTreeClassifier(min_samples_split= min_samples_split, random_state= RANDOM_STATE )
    model.fit( X_train, y_train)
    
    predictions_train = model.predict(X_train)
    predictions_val = model.predict(X_val)
    
    accuracy_train = accuracy_score(predictions_train, y_train)
    accuracy_val = accuracy_score(predictions_val, y_val)
    
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
    
plt.title('Train X validation metrics')
plt.xlabel('min_samples_split')
plt.ylabel('accuracy')
plt.xticks(ticks=range(len(min_samples_split_list)), labels=min_samples_split_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train', 'Validation'])
```

- Now how increasing the number of min_samples_split reduce overfitting
    - Increasing min_samples_split from 10 to 30, and from 30 to 50, even though it does not improve the validation accuracy, it brings the training accuracy closer to it, showing a reduction in overfitting.

```python
accuracy_list_train = []
accuracy_list_val = []
for max_depth in max_depth_list:
    # We can fit the model at the same time we define it, because the fit function returns the fitted estimator
    model = DecisionTreeClassifier(max_depth=max_depth, random_state= RANDOM_STATE)
    model.fit(X_train, y_train)
    
    predictions_train = model.predict(X_train) ## The predicted values for the train dataset
    predictions_val = model.predict(X_val) ## The predicted values for the test dataset
    accuracy_train = accuracy_score(predictions_train,y_train)
    accuracy_val = accuracy_score(predictions_val,y_val)
    accuracy_list_train.append(accuracy_train)
    accuracy_list_val.append(accuracy_val)
    
plt.title('Train x Validation metrics')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.xticks(ticks = range(len(max_depth_list )),labels=max_depth_list)
plt.plot(accuracy_list_train)
plt.plot(accuracy_list_val)
plt.legend(['Train','Validation'])
```

- Reducing max_depth can help to reduce overfitting
    - Reducing max_depth from 8 to 4 increases validation accuracy closer to training accuracy, while significantly reducing training accurary.
    - The validation accuracy reaches the highest at tree_depth = 4.
    - When the max_depth is smaller than 3, both training and validation accuracy decreases. The tree cannot make enough splits to distingish positives from negatives (the model is underfitting the training set)
    - When the max_depth is too high (>=5), validation accuracy decreases while training accuracy increases, indicating 
- So we can choose the best values for these 3 hyperparameters for our model to be:
    max_depth = 4
    min_samples_split = 50

```python
decision_tree_model = DecisionTreeClassifier(min_samples_split=50, max_depth=3, random_state= RANDOM_STATE)
decision_tree_model.fit(X_train, y_train)
```

```python
print(f"Metrics train:\n\tAccuracy score: {accuracy_score(decision_tree_model.predict(X_train),y_train):.4f}")
print(f"Metrics validation:\n\tAccuracy score: {accuracy_score(decision_tree_model.predict(X_val),y_val):.4f}")
```

- No overfitting, even though the metrics are not that good
