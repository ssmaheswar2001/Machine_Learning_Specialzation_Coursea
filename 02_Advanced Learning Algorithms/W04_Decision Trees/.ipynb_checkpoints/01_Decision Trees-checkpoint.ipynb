# Decision Trees
- 
- The model that is output by learning algorithm looks like a tree.
- 
- 
- The job of the decision tree learning algorithm is, out of all possible decision trees, to try to pick one that hopefully does well to new data such as our cross-validation and test sets as well.

# Learning Process
- The process of building a decision tree given a training set has a few steps.
1. We have to decide what feature to use at the root node via an algorithm.
2. Focusing jut on the left part or sometimes called the left branch of the decision tree to decide what nodes to put over there. In particular, what features that we want to split on or what feature do we want to use next via an algorithm.
- 
- Through the proceess of build the decision tree, there were couple of key decisions that we had to make at various steps during the algorithm
    1. How do we choose what features to use to split on at each node? The decision trees will choose what feature to split on in order to try to maximize purity.
    The decision tree learning algorithm has to choose b/w which of these features results in the greatest purity of the labels on the left and right sub branch.
    2. To decide when do we stop splitting.
        - When a node is 100% one class.
        - We might also decide to stop splitting and no further results in the tree exceeding the maximum depth, where the maximum depth that we allow the tree to go to, is a parameter that we could just say. In decision tree, the depth of the node is defined as the number of hops that it takes to get from the root node that is denoted the very top to that particular node. So the root node takes zero hops, to get to itself and is at Depth 0. We want to limit the depth of the decision tree is to make it less prone to overitting.
    3. When the improvement in purity score are below a threshold. If the splitting a node results in minimum improvements to purity or we see later is acutally decreases impurity
    4. If the number of examples that a node is below a certain threshold, then we might also decide to stop splitting.
