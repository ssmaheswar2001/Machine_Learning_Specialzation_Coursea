# What is clustering
- A Clustering algorithm looks at a number of data points and automatically finds data points that are related or similar to each other.
- In supervised learning, the dataset included both the inputs x as well as the target outputs y.
    
- In unsupervied learning, we are given a dataset with just x, but not the lables or the target labels y.
    
- Because we don't have target labels y, we're not able to tell the algorithm what is the 'right answer y' that we wanted to predict. Instead we're going to ask the algorithm to find someting intresting about the data, that is to find some intresting structure about this data.
- Clustering algorithm which looks for one particular type of structure about the data.
    
- Applications of clustering,
    
- The most commonly used clustering algorithm called the k-mean algorithm.

# K-means Algorithm
- To randomly initialize K cluster centriods, Mu 1 Mu2 through Muk. In our examples, K = 2.

- There is one corner case of the algorithm, which is what happens if a cluster has zero training examples assigned to it. In that case, the 2nd step, Mu k would be trying to compute the average of zero points. If that ever happens, the most common thing to do is to just eliminate that cluster. So we end up with K-1 clusters. Or if we really need K clusters an alternative would be just randomly reinitialize that cluster centroid and hope that is gets assigned atleast some points next time round.
    
- It turns out that K-means is also frequently applied to data sets where the clusters are not that well seperated. If we were to run K-means with say, 3 cluster centriods, we might find that k-means would group these points into 3 clusters. This K-means working just fine and giving a useful results even if the data doesnot lie in well-spearate groups or clusters.
    
- Why we might hope this algorithm does converge? K-means trying to optimize a specific cost function.

# Optimization Objective
- In supervised learning algorithm, we have taken training set posing a cost funtion, and then using gradient descent or some other algorithms to optiize that cost function.
- What the k means algorithm is doing is trying to find assignments of points of clusters centroid as well as find locations of clusters centroid that minimized the squared distance.
    
- If we were to compute the cost function, it would be like everyone at the blue points and measure the distances and compute square. And similarly look at everyone of the redpoints and compute these distances and compute the square. And then average of the squares if all of these differences for the red and the blue points is the value of the cost fucntion J. Inorder to reduce the cost function J, on every step is to try to update the cluster assignments C1 through C30 or udpate the positions of the cluster centers.
    
    - This function is called the distortion cost function.

- The distortion cost function should go down or stay the same, but if it ever fails to go down ot stay the same, then there is a bug in the code, it should never go up

# Initializing K-means
- The very 1st step of the k-means clustering algorithm, was to choose random locations as the initial guesses for the cluster centroids mu_1 through mu_k.
- 
- When running k-means we should always choose the number of cluster centroids K to be lessened to training examples m.
- Inorder to choose the cluster centroids, the most common way is to randomly pick k training examples.
- Depending on how we choose the random initial central centroids K-means will end up picking a difference set of causes for our data set.
      
- If we want to give k means multiple shots at finding the best local optimum. If we want to try multiple random initilization, so give it a better chance of finding good clustering. We could do with the k-means algorithm is to run  it multiple times and then try to find the best local optima. Then one way to choose the best between these solutions is to compute the cost function J and pick one which has the lowest value foer the cost function J.

# Choosing the number of clusters
- For a lot of clustering problems, the right value of k is truly ambiguous.
    
- The one way to choose the value of K is called the elbow method and what that does is we would run K-mean with a variety of K valus and plot the cost function or distortion function J as a function of  the number of clusters.
- For one cluster, the distortion function or the cost function J will be very high and as we increase the number of clusters, it will go down, maybe as follows, and if the curve say like the cost function is decreasing rapidly until we get to 3 clusters but the decrease is more slowly after that. So we choose K =3 and this is called elbow./
- But for lot of applications the cost function decreases smoothly and it doesn't have a clear elbow so we could not pick K.
- Choosing K to minimize the cost function J is not a good technique.
- 
- We can choose the K value, by evaluating the K-means based on how well it performs for the later purpose.
