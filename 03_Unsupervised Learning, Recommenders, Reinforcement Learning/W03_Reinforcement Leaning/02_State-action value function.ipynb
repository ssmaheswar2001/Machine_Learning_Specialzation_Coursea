# State-action value function definition (Q function)
- A key quality that reinforcment learning algorithm will try to compute and that's called the state action value function.
- The state action value function is a function typically denoted by the uppercase Q. It's a function of a state we might be in as well as the action we might choose to take in that state and Q(s,a). Will give a number that equals the return. 
- If we start in the state S and take the action A just once and after taking ation A once we then behave optimally after that. So after that we take whatever actions will result in the highest possible return.
- How do we know what is the optimal behavior? And if we knew what the optimal behavior, if we already knew what's the best action to take in every state, why do we still need to compute Q(s,a). Because we already have the optimal policy
- Q(s, a) will be equal to the total return if we start from say that take the action A and then behave optimally after that. Meaning take actions according to the policy.
- Q function or state-action function tells us what are our returns or really what is tha value? How good is it? Just take action A and state S and then behave optimally after that. Now it turns out that once we can compute the Q function this will give us a way to pick actions as well.  
- It turns out that the best possible return from any state S. Is the largest value of Q(s,a), maximizing over A. So the best possible action status is the action A. That actually maximizes Q(s, a)
- S can just pick the action A that gives the largest value of Q(s, a) and that will turn out to be a good action. In fact it turned out to be the optimal action. Q(s, a) is returned if we start in the state S and take the action A and then behave optimally after that. So inorder to earn the biggest possible return, what we really want is to take the action A. That results in the biggest total return.
- 
- To summarize, if we compute Q(s, a), for every state and every action, then that gives us a good way to compute the optimal policy pi(s). So that's the state action value function or the Q function.

# Random (stochastic) environment
- In some applications, when we take an action, the outcome is not always completely reliable.
- 
- The reinforcement learning problem is stochastic, there isn't one sequence of rewards that we see for sure instead we see the sequence of different rewards. In a stochastic reinforcement learning problem, what we're intrested in is not maximizing the return because that's random number. What we're intrested in is maximixing the average value of the sum of discounted rewards. This means we want to maximize what we expected to get on average in terms of the sum of discounted rewards.
- 
- The job of the reinforcement learning algorithm is to choose a policy Pi to maximize the average or the expected sum of discounted rewards.
- To summarize, when we have a stochastic reinforcement learning problem or a stochastic Markov decision process the goal is to choose a policy to tell us what action to take in state S so as to maximize the expected return.
