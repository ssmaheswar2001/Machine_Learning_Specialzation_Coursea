# Algorithm refinement : Improved neural network architecture
- A neural network architecture that will input the state and action and attempt to ouput the Q function, Q(s, a).
- The neural network architecture that we saw previously, where it would input 12 numbers and output Q(s, a). Whenever we are in some state s, we would have to carry out inference in the input network separately 4 times to compute these 4 values so as to pick the action a that gives us the largest Q value. This is inefficent because we have to carry our inference 4 times from every single state. 
    
- Instead, it turns out to be more efficient to train a single network to output all 4 of these values simultaneously
- The modified neural network, the job os the neural network is to compute simultaneosly the Q value for all 4 possible actions for when we are in the state s. This turns out to be more efficient becuase gien the state s we can run inference just once and get all 4 of these  values, and then very quickly pick the action a that maximizes Q(s, a)

# Mini-batching and soft-updates 
- The 2 refinement to the reinforcement learning algorithm.
- The first idea is called mini-batches, and this turns out to be an idea they can both speedup our reinforcement learning algorithm and it's also applicable to supervised learning. They can help us speed up our supervised learning algorithms as well, like training a neural network, or training a linear regression, or logistic regression model.
- The second idea we'll look at is soft updates, which it turns out will help our reiforcement learning algorithm do a better job to converge to good solution.
- To understand mini-batches, let's just look at supervised learning to start. When the training set size is very large, the gradient descent algorithm turns out to be quite slow. The idea of mini-batch gradient descent is to not use all million traing examples on every single iterative through the loop. Insetad we may pick a smaller number, let us call it m prime equals say 1000. On every step, instead of all million training examples, we would pick some subset of 1000 or m prime examples. Now each iteration through gradient descent requires looking only at 1000 rather millions examples, and every step takes much less time and just leads to a more efficent algorithm
    
    
    
- To our reinforcement learning algorithm, even if we stored all 10000 most recent tuples in the replay buffer, what we might choose to do is not use all 10000 every time we train a model. Instead, what we might do is just take the subset. We might choose just 1000 examples of these (s, a, R(s), s^1) tuples and use it to create just 1000 training examples to train the neural network. It turns out that the overall tend to speed up the reinforcemetn learning algorithm.
    
- Finally, there's one other refinement to the algorithm that can make it converge more reliably.
- The soft update method helps to prevent Q_new through just one unlucky step getting wrose.
- This is called a soft update because whenever we train a new neural network W_new, we're only going to accept a little bit of the new value. These numbers 0.001 and 0.99, these are hyperparameters that we could set, but it controls how aggresively we move W to W-new and thse 2 numbers are expected to add up to 1.
    
- The soft update allows us to make a more gradual change to Q or to the neural network parameters W and B that affect our current guess for Q function Q(s, a). It turns out that using soft update method causes the reinforcement learning algorithm to converge more reliably.
