# The return in Reinforcement Learning
- What are the states of reinforcement learning application, as well as how depending on the actions we take us go through different states, and also get to enjoy different rewards. But how do we know if a particular set of rewards is better or worse than a different set of rewards? The return in reinforcement learning allows us to capture that.
- The return is defined as the sum of the rewards but weighted by one additional factor, which is called the discount factor. The discount factor is a number a little bit less than 1. 
- In many reinforcement learning algorithms, a common choice for the discount factor will be a number pretty close to 1, like 0.99 or 0.99 or 0.999. 
- This very heavily down weights or very heavily we say discounts rewards in the furture, because with every additional passing timestamp, we get only half as much rewards that we would have gotten one step earlier.
- 
- The return we get depends on the rewards, and the rewards depends on the actions we take, and so the return depends on the actions we take.
- 
- The return in reinforcement learning is the sum of the rewards that the system gets, weighted by the discount factor, where rewards in the far future are weighted by the discount factor raised to a higher power.
