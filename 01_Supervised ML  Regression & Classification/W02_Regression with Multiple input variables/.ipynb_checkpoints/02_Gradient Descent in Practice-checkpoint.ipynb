# Feature Scaling
- The technique called feature scaling that will enable gradient descent to run much faster
- Relationship b/w size of a feature (number of features) and the size of its associated parameters\
- 
- We may notice that when a possible range of values of a feature is large, it's more likely that a good model will learn to choose a relatively small parameter value. Likewise, when the possible values of the feature are small, then a reasonable value for its parameters will be relatively large
- How does this relat to gradient descent?
- Look at the the scatter plot of features and we can notice that the horizontal axis is on a much large scale or much large range of values compared to the vertical axis.
  Similarly, look at the cost function might look in a contour plot, and you might see a contour lot where the horizontal axis has amuch narrow range ad whereas the vertical axis takes on much larger values.
  
- This is because a very small change to w1 can have a very large impact on the estimated price and that's a very large impact on the cost J. Becuase w1 is tends to be multiplied by a very large number. In contrast, it takes a much larger change in w2 inorder to change the predictions much. And thus small changes to w2, don't change the cost function nearly as much. 
- The contours are so tall and skinny gradient descent may end up bouncing back and forth for a long time before it can finally find its way to the global minimum.
    
- A useful thing to do is to scale the features. This means performing some transformation of your training data so that x1 may range from 0 to 1 and x2 also range from 0 to 1. So the data points looks as
   
- The key point is that the rescale x1 and x2 are both now taking comparable ranges of values to each other. So the gradient descent on a cost function to the tranformed data can look more lik circles and now gradient descent can find a much more direct path to the global minimum.
    


- One way to get a scale version it to take each original value and divide by the maximum of the range.
    
- In addtion to dividing by the maximum, we can also do mean normalization
#### Mean Normalization
- We start with the original features and then we re-scale them so that both of them are centered around zero. 
- Before they only had values greater than zero, now they have both negative and positive values that may be usually between -1 and +1.
- To calculate the mean normalization, first find the average, also called the mean on the training set 
    
#### Z- score Normalization
- There is one common re-scaling method called Z-score normalization.
- To implement Z-score normalization, we need to calculate mean and standard deviation of each feature

## Checking Gradient descent for convergence
- When running gradient descent, how can we tell if it is converging?
- The gradient descent rule,
    
- The job of the gradient descent is to find parameters w and b that hopefully minimize the cost Function J, which is calculated on the training set
- It the gradient descent is working properly, thenn the cost J should decrease after every single iteration. If J ever increases after one iteration, that means either Alpha is choosen poorly, and it usally means Alpha is too large
- Another way to decide when the model is done training is with an automatic convergence test.
- If the cost J decrease by less than this epsilon number on one iteration,, then we're likely on this flattend part of the curve and we can declare convergence
- 

### Choosing the Learning Rate
- Our learning algorithm will run much better with an appropriate choice of learning rate.
- If it's too small, it will run very slowly
  If it is too large, it may not even converge.
- If we plot the cost for a number of iterations and notice that the costs sometimes goes up and sometimes goes down, we should take that as a clear sign that gradient descent is not working properly. This could mean that our learning rate is too large.
- To fix this we use a smaller learning rate, then our updates goes down, and we'll consistently decrease until it reaches the global minimum.
    
- But, setting Alpha to be really small is not going to be the most efficeitn choice for actually training our learning algorithm. If the learning rate is too small, then gradient descents can take a lot of iterations to converge.
- When running gradient descent, try a range of values for the learning rate Alpha. 
- Try a range of values until I found the value of that's too small andd then also make sure we've found a value that's too large.

## Feature Scaling

- Rescaling the dataset so the features have a similar range.
- Feature scaling, essentially dividing each postive feature by its maximum value, or more generally, rescale each feature by both its minimum and maximum values using (x-min)/(max-min). Both ways normalizes features to the range of -1 and 1, where the former method works for positive features which is simple and serves well.
- Mean Normalization
    
- Z-score normalization
#### Z-score normalization
- After z-score normalization, all features will have a mean of 0 and a standard deviation of 1.
- To implement Z-score normalization, adjust the input values as shown
    
- Where j selects a feature or a column in the X matrix. 
    

```python
def zscore_normalize_features(X):
    '''Compute X, zscore normalized by column
    
    Args : 
        X (ndarray (m,n)) : Input data, m examples, n features
        
    Returns : 
        X_norm (ndarray (m, n)) : Input normalized by column
        mu (ndarray (n,)) : Mean of each features
        sigma (ndarray (n,)) : Standard deviation of each feature
    '''
    
    # Find the mean of each column / feature
    mu = np.mean(X, axis = 0)
    # Find the standard deviation of each column / feature\
    sigma = np.std(X, axis = 0)
    X_norm = (X - mu) / sigma
    
    return (X_norm, mu, sigma)
    
```

## Feature Engineering

## Polynomial Regression
- Let's take the ideas of multiple linear regresson adn feature engineering to come up with a new algorithm called polynomial regression., which will let us fit curves, non-linear functions, to our data.
- Let's say we have a housing data-set. It doesn't look like a straight line fits this data-set-set very well. Maybe we want to fit a curve, may be a quadratic function to the data. But quadratic model doesn't really make sense becuase a quadratic function eventually comes back down
    
- So, now we may choose a cubic function. This model produces the curve, which is a somewhat a better fit to the data becuase the size does eventually come back up as the size increases. 
    
- These are both examples of poynomial regression.
- If we create features that are these powers like square of the original features like this, then feature scaling becomes increasingly important.
- If the size of the house ranges from say 1-1000, then the second features (size square) ranges from one to a million, and the third feature, which is size cubed, ranges from one to a billion.
- If we're using gradient descent, it's important to apply feature scaling to get the features into comparable ranges of values.
- By using feature engineering and polynomial functions, we can potentially get a much better model for our data.

### Polynomial Features
- Consider a scenario where the data was non-linear. 

```python
def compute_gradient(X, y, w, b):
    '''
    Compute the gradient for linear regression
    Args : 
    X (ndarray (m, n)) : Data, m examples with n features
    y (ndarray (m,)) : target values
    w (ndarray (n, )) : model parameters
    b (scalar) : model parameters
    
    Returns:
    dj_dw (ndarray (n, )) : The graident of the cost w.r.t the parmeters w
    dj_db (scalar): The gradient of the cost w.r.t the parameters b
    '''
    
    m, n = X.shape
    dj_dw = np.zeros((n, ))
    dj_db = 0
    
    for i in range(m):
        err = (np.dot(X[i], w) + b) - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err * X[i, j]
        dj_db = dj_db + err
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_db, dj_dw
```

```python
def cost_function(X, y, w, b):
    '''
    Compute cost
    Args
    X (ndarray (m,n)) : Data, m examples with n features
    y (ndarray (m,)) : Target values
    w (ndarray (n, )) : Model paramaters
    b (scalar) : Model parameters
    
    Returns
    cost (scalar) : cost
    '''
    m = X.shape[0]
    cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b  # (n, )(n, ) = scalar 
        cost = cost + (f_wb_i - y[i])**2 # scalar
    cost = cost / (2*m)
    return cost
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iters): 
    """
    Performs batch gradient descent to learn w and b. Updates w and b by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      X (ndarray (m,n))   : Data, m examples with n features
      y (ndarray (m,))    : target values
      w_in (ndarray (n,)) : initial model parameters  
      b_in (scalar)       : initial model parameter
      cost_function       : function to compute cost
      gradient_function   : function to compute the gradient
      alpha (float)       : Learning rate
      num_iters (int)     : number of iterations to run gradient descent
      
    Returns:
      w (ndarray (n,)) : Updated values of parameters 
      b (scalar)       : Updated value of parameter 
      """
    
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = compute_gradient(X, y, w, b)   ##None

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               ##None
        b = b - alpha * dj_db               ##None
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```

```python
# Create target data
x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1, 1)

intial_w = np.zeros_like([0])
intial_b = 0
model_w, model_b,J_hist = gradient_descent(X, y, intial_w, intial_b, 1e-2, 1000)    

plt.scatter(x, y, marker='x', c='r', label = 'Actual Value')
plt.title("No feature engineering")
plt.plot(X, X*model_w + model_b, label='Predicted Value')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

- Well, as expected not a greate fit. We need y = wx2 + b or a polynomial feature. To accomplish this, we can mofiy the input data to engineer the needed features. If we swap the original data with a version that squares the x value, then we can achieve.

```python
# Create target data
x = np.arange(0, 20, 1)
y = 1 + x**2

# Engineer features
X = x**2
X = X.reshape(-1, 1)
```

```python
intial_w = np.zeros_like([0])
intial_b = 0
model_w, model_b,J_hist = gradient_descent(X, y, intial_w, intial_b, 1e-5, 1000)  
plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("Added x**2 feature")
plt.plot(x, np.dot(X,model_w) + model_b, label="Predicted Value"); plt.xlabel("x"); plt.ylabel("y"); plt.legend(); plt.show()
```

### Selecting Features
- We knew that x**2 term was required. It may not always be abvious which features are required. One could add a variety of potential features to ty and find the most useful
    

```python
# Create target data
x = np.arange(0, 20, 1)
y = x**2

# Engineer features
X = np.c_[x, x**2, x**3]
```

```python
intial_w = np.zeros_like([0, 0, 0])
intial_b = 0

model_w, model_b,J_hist = gradient_descent(X, y, intial_w, intial_b, 1e-7, 10000)  
print("w = ",model_w," b = ",model_b)
plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("x, x**2, x**3 features")
plt.plot(x, X@model_w + model_b, label="Predicted Value"); plt.xlabel("x"); plt.ylabel("y"); plt.legend(); plt.show()
```

- This implies the model after fitting/training is
    
- The gradient descent has emphasized the data that is the best fit to the x**2 data by increasing w1 term relative to the others. 
- Gradient descent is picking the 'correct' features for us by emphasizing its associated parameters
- The idea is : 
    - Less weight value implies less important/correct feature, and in extreme. when weight becomes zero or very close to zero, the associated feature is not fitting the model to the data.
    - After fitting, the weight associated with the x**2 feature is much larger than the weights for x or x**3 as it is most useful in fitting the data.
