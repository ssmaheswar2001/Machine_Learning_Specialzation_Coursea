# What is ML?
- Field of study that gives computers the ability to learn without being explicitly programmed.

- There are 2 main types of ML algorithms
1. Supervised learning : Used most in real world applications
2. Unsupervised learning 

## Supervised Learning
- Supervised learning, refers to algorithms that learn x to y or input to output mappings.

- The key characterisitic of supervised learning is that you give your learning algorithm examples to learn from. That includes the right answers, means, the correct label y for a given x, and is by seeing the correct pairs of input x and desired output label y that the learning algorithm eventually learns to take just the input alone without the output label and gives a resonably acccurate prediciton or guess the output.

- In all these applications, we will first train your model with examples of inputs x and the right answers, that is the labels y.
- After the model has learned from these input, output, or x and y pairs, they can then take a brand new input x, something it has never seen before, and try to produce the appropriate corresponding output y.

### Supervised Learning Types
- There are 2 major types of supervised ML algorithms
1. Regression : To Predict numbers out of infinitely many possible numbers
    - Want to predict housing prices based on the size of the house. You've collected some data and say you plot the data and it looks like this. 
    
    - Here on the horizontal axis is the size of the house in square feet. Here on the vertical axis is the price of the house in, say, thousands of dollars. 
    - We're trying to predict a number from infinitely many possible numbers such as the house prices
    - Here the task is to predict number.
    
2. classification
    - Let's take a breast cancer detection.
    - We will designate in this example with a 0 or maligant, which will designate with a 1.
    - In this case only 2 possible outputs 0 or 1, benign or malignant.
    
    - This is different from regression which tries to predict any number, all of the infinitely many number of possible numbers.
    
    - Because there are only 2 possible outputs or 2 possible categories
    - Classification problems can even have more than 2 categories
- In a regression, the learning algorithm has to predict numbers from infinitely many possible output numbers.
- Whereas in classification, the learning algorithm has to make a prediction of a category, all of a small set of possible outputs.
