### Sigmoid or Logistic Function
- For a classification task, we can start by using our linear regression mode, to predict y given x.
- However, we would like the predictions of our classification model to be b/w 0 and 1 since out output variable y is either 0 or 1. 
- This can be accomplished by using a 'sigmoid function' which maps all input valuess to values b/w 0 and 1.
#### Formula for Sigmoid function
- The formula for a sigmoid function is as follows : 
    
- In the case of llogistic regression, z (the input to the sigmoid fucntion), is the output of a linear regression model.
    
- In the case of a single example, z is a scalar.
- In the case of multiple examples, z may be a vector consisting of m values, one for each example.
- The implementatoons of the sigmoid function should cover both these potential inputs formulas.

- NumPy has a function called exp(), which offers a convenient way to calculate the exponential (e^z) of all elements in the input array (z).

```python
# Input is a n array
input_array = np.array([1, 2, 3])
exp_array = np.exp(input_array)

print("Input to exp : ", input_array)
print("Output of exp : ", exp_array)

# Input is a single number
input_val = 1
exp_val = np.exp(input_val)

print("Input to exp:", input_val)
print("Output of exp:", exp_val)
```

```python
def sigmoid(z):
    '''compute the sigmoid of z
    
    Args : 
        z (ndarray) : A Scalar, numpy of any size.
        
    Retrurns : 
        g (ndarray) : sigmoid(z), with the same shape as z
    '''
    
    g = 1/(1+np.exp(-z))
    return g
```

```python
# Generate an array if evenly spaced values b/w -10 and 10
z_tmp = np.arange(-10, 11)

# Used the function implemented above to get the sigmoid values
y = sigmoid(z_tmp)

print("Input (z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
# Plot z vs sigmoid(z)

fig, ax = plt.subplots(1, 1, figsize=(5, 3))
ax.plot(z_tmp, y, c='b')

ax.set_title('Sigmoid Function')
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

As the sigmoid fnction appraocehs 0 as z goes to large negative values and approaches 1 as z goes to large positive values

# Decision Boundary
- What if we want to learn the algorithm to predict. Is the value of y going to be zero or one?
- A common choice would be pick a threshold of 0.5 so that if f(x) is greater than or equal to 0.5, then predict y is one. or if f(x) is less than 0.5, then predict y is 0.
- g(z) is greater than or equal to 0.5 whenever z is greater than or equal to 0. That is whenever z is on the right half of the axis. Conversely, when z is less than zero, the algorithm predicts y is 0.
- 
- Let's take an example of classification problem where we have 2 features, x1 and x2 instead of just 1 feature. 
- Decision boundary is called when z = 0 becuase that's the line where we're just almost neutral about whether y is 0 or y is 1.
- The line turn out to be the decision boundary, where if the feature x are to the right of decision boundary line, logistic regression predicts 0.

- Lets look at a more complex example where the decision boundary is no longer a straight line. 
- This is implementing polynomial linear regression to the logistic regression. With the chice of features, polyomial features into a logistic regression.
- Here the decision boundary line turns out to be circle.
- 
- We can even come with more complex decision boundaries than these, by having higher-order polynomial terms. 
- 
- By this, logistic regression can learn to fit complex data
